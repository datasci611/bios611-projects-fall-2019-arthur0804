```python
import pandas as pd
```

### Data about clients

```python
# demographic data
client_demo = pd.read_csv("../data/raw/CLIENT_191102.tsv", sep = "\t")
# entry and exit data
record = pd.read_csv("../data/raw/ENTRY_EXIT_191102.tsv", sep = "\t")
```

```python
client_demo.head()
```

```python
record.head()
```

```python
# select columns of interest
client_demo.drop(["Client Unique ID"], inplace=True, axis = 1)
record.drop(["Client Unique ID", "Entry Exit Group Id", "Entry Exit Household Id", "Unnamed: 6", "Housing Move-in Date(5584)",
                             "Entry Exit Type", "Entry Exit Date Added", "Entry Exit Date Updated"], inplace=True,  axis = 1)
```

```python
# merge two dataframes
client_data_merged = pd.merge(client_demo, record, how="inner", 
                              left_on=["EE UID", "Client ID", "EE Provider ID"], right_on=["EE UID", "Client ID", "EE Provider ID"])
```

```python
# parse the date
client_data_merged['Entry Date'] = pd.to_datetime(client_data_merged['Entry Date'].astype(str), format='%m/%d/%Y')
client_data_merged['Exit Date'] = pd.to_datetime(client_data_merged['Exit Date'].astype(str), format='%m/%d/%Y')
```

```python
# add one column of stay length
client_data_merged["length of stay"] = (client_data_merged['Exit Date'] - client_data_merged['Entry Date']).dt.days
```

```python
# save
client_data_merged.to_csv("../data/cleaned/client_data_merged.tsv", index=False, sep = "\t")
```
